```
from requests_html import HTMLSession
```

```
session=HTMLSession()
```

```
def kobita_scraping(kobita,url,total_page):
  for i in range(1,total_page+1):
    for j in poet:
      poem_url=url+f"/{j}/?t=p&pp="+str(i)
      print(poem_url)
      r=session.get(poem_url)

      poem_name=".col-sm-12"
      poem_des="p"
      poems_name=[i.text for i in r.html.find(poem_name)]
      poems_des=[i.text for i in r.html.find(poem_des)]
      with open(f"{j}/{j}_poemlist_{i}.txt","w") as f:
        f.write("poem_list\n")
        for t in zip(poems_name):
          f.write(f"{t}")
          f.write('\n')
      print(poems_name)
poet=("rabindranath","shamsurrahman","nazrulislam","jibanananda","sukumar")
poem_url="https://www.bangla-kobita.com"
pageno=29
kobita_scraping(poet,poem_url,pageno)
```
